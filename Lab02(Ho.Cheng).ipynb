```python
#Part2
#1.
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
t = pd.read_csv('titanicSurvival_m.csv')

#2.
print(t.isnull().sum())

#3.
print(t.describe())

#4.
plt.scatter(t.Age.fillna(0),t.Fare.fillna(0))

#5.
plt.scatter(t.Age.fillna(t.Age.mean()),t.Fare.fillna(t.Fare.mean()))

#6.
#From the results,for the scatter plot which missing values were replaced by 0, 
#the replaced values are all concentrated along the 0 value of x(Age) and y(Fare) axis,
#which is quite obvious on the plot. Another scatter plot which missing value replaced
#by mean have less obvious on this issue.
```
